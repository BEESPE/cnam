A partir du travail dj effectu, nous allons afficher sur la mme figure plt les histogrammes superposs des pixels des deux images (en utilisant deux couleurs, par exemple rouge et bleu). Il est possible de jouer sur le paramtre alpha (transparence) de la fonction hist afin de bien visualiser les deux histogrammes.

```python
def plot_two_hists(
    origin_img_array: np.array,
    filtered_img_array: np.array,
    filter_name: str,
) -> None:
    plt.hist(
        origin_img_array,
        bins=20,
        density=True,
        alpha=0.5,
        edgecolor="black",
        color="red",
        label="Image originale",
    )
    plt.hist(
        filtered_img_array,
        bins=20,
        density=True,
        alpha=0.5,
        edgecolor="black",
        color="blue",
        label=f"Filtre {filter_name}",
    )
    plt.legend()
    plt.show()

plot_two_hists(
    origin_img_array=img_array.flatten(),
    filtered_img_array=np.array(img_blur).flatten(),
    filter_name="BLUR",
)
```

Il est dsormais temps de transformer l'image en nuance de gris. Il existe plusieurs faons de le faire, mais une faon simple est de prendre la moyenne pondre des valeurs RGB de l'image originale.

Ecrire une fonction permettant de gnrer une image en niveaux de gris  partir d'une image originale RGB (sous forme np.array) en pondrant chacun de ses trois canaux de couleur. L'appeler sur notre image en testant diffrentes configurations de poids pour visualiser leur effet sur limage en nuance de gris gnre.

```python
def convert_to_greyscale(img_rgb_array: np.array) -> np.array:
    return (img_rgb_array[:, :, 0] + img_rgb_array[:, :, 1] + img_rgb_array[:, :, 2]) // 3
```

```python
img_grey_array = convert_to_greyscale(img_rgb_array=img_array)
```
